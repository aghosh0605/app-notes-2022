**Python program to display all the prime numbers within an interval**

```python
def isPrime(a):
    value=True
    if(a==0 or a==1):
            value=False
            return value
    else:
        for i in range (2,(int(a/2)+1)):
                if(a%i==0):
                    value=False
        return value
        
    
if __name__=="__main__":
    num1=int(input("Enter 1st interval: "))
    num2=int(input("Enter 2nd interval: "))
    print("Prime Numbers are ", end="")
    for i in range(num1,num2):
            result=isPrime(i)
            if result==True:
              print(str(i)+" , " ,end="")
```

**Python program to check if the number is an Armstrong number or not**

```python
num=int(input("Enter number: "))
num1=num
str1=str(num)
lenth=len(str1)
total=0
while num>0:
    temp=num%10
    total=total+temp**lenth
    num=num//10
if(total==num1):
    print("Armstrong number")
else:
    print("Not Armstrong number")
```

**Program to check Armstrong numbers in a certain interval**

```python
def isArmstrong(num):
    num1=num
    str1=str(num)
    lenth=len(str1)
    total=0
    while num>0:
            temp=num%10
            total=total+temp**lenth
            num=num//10
    if(total==num1):
            return True
    else:
            return False

if __name__== "__main__":
    num1=int(input("Enter 1st number: "))
    num2=int(input("Enter 2nd number: "))
    print("Armstrong Numbers are ")
    for i in range(num1,num2):
            result=isArmstrong(i)
            if(result):
              print(str(i)+" , ",end="")
            
```
