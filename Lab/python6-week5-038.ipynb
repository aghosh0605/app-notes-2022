1.1 **Write a query to display the total number of recipes available with the description “Chinese”**

```python
def get_data(des):
    try:
      count = 0
      statement = "SELECT id,name FROM recipes WHERE description=%s"
      data = (des,) #it must be of type list, tuple or dict
      cursor.execute(statement, data)
      for (id,name) in cursor:
        count+=1
        print(f"Successfully retrieved {id}: {name} ")
      print("Total Number:",count)
    except database.Error as e:
      print(f"Error retrieving entry from database: {e}")
```
