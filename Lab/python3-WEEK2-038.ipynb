1. **Given a string, find its mirroring image**

````python
mir_dict = {'b':'d', 'd':'b', 'i':'i', 'o':'o', 'v':'v', 'w':'w', 'x':'x'}
res = ''
str1 = input("Enter string to mirror: ")
for i in str1:
  if i in mir_dict:
    res+=mir_dict[i]
  else:
    print("Not Possible")
    break
print(res)
````

8. **Write a Python program to Convert Tuple Matrix to Tuple List**
```
Sample Input : [[(9, 51), (7, 9)], [(11, 1), (22, 19)]]
Output : [(9, 7, 11, 22), (51, 9, 1, 19)]
```



````python
list1 = [[(9, 51), (7, 9)], [(11, 1), (22, 19)]]
temp = [ele for sub in list1 for ele in sub]
list2 = list(zip(*temp))
print(list2)
````

12.	**Write a Python function (using function) that checks whether a passed string is palindrome or not**
Note : function type - No arguments with return values

````python
def checkPalin():
  pal = input("Enter a sequence of input: ")
  count=0
  for i in range(0,len(pal)):
    if not (pal[i]==pal[len(pal)-1-i]):
      break
    count+=1
  if count==len(pal):
    return True
  else:
    return False
if __name__=='__main__':
  result=checkPalin()
  print(result)
````

13. **Write a Python function (using function) that checks whether a given number is prime or not.**
Note : function type - with arguments with return values

````python
def checkPrime(num):
  for i in range(2,(num//2)+1):
    if num%i==0:
      return False
      break
  return True

if __name__=='__main__':
  num = int(input("Enter number to check: "))
  result = checkPrime(num)
  print(result)
````
